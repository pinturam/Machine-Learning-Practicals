### 20. Invert Matrix

The inverse of a square matrix, A, is a second matrix A–1, such that:
AA(power−1) = I where I is a Identity matrix.

```python
# laod library
import numpy as np
```

```python
# create a square matrix
m = np.array([[2,3],
              [2,4]])
```

```python
# inverse matrix
np.linalg.inv(m)
```

```python
# identity matrix
m @ np.linalg.inv(m)
```
